```python
# HeH+ molecule data
Delta = 0.0
N = 3
R = 1.4
Zeta1 = 1.24
Zeta2 = 1.24
Za = 1.0
Zb = 1.0
hf = HartreeFock(N,R,Zeta1,Zeta2,Za,Zb)

print("Converged electronic energy:", hf.Energy)
print("Converged total energy:",hf.Energytot)
print("Energy eigenvalues",hf.epsilon)
print("Matrix coefficients",hf.C)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fmin

# define functions to calculate energy and total energy
def energy(R):
    hf = HartreeFock(N, R, Zeta1, Zeta2, Za, Zb)
    return hf.Energy

def total_energy(R):
    hf = HartreeFock(N, R, Zeta1, Zeta2, Za, Zb)
    return hf.Energytot

# vectorize the functions to operate on arrays
energy = np.vectorize(energy)
total_energy = np.vectorize(total_energy)

# define range of R values to plot
RR = np.linspace(0.1, 6., 100) 

# calculate equilibrium distance and minimum energy
R_eq = fmin(lambda R: total_energy(R), 0.01, disp=False)[0]
min_energy = total_energy(R_eq)

# print results

print("Equilibrium distance from input = {:.2f}".format(R))
print("Total Energy = {:.4f}".format(total_energy(R)))
print("Equilibrium distance calculated from the graph = {:.2f}".format(R_eq))
print("Minimum Energy = {:.4f}".format(min_energy))

# plot the energy curve and equilibrium distance
plt.title("HeH+ molecule")
plt.plot(RR, total_energy(RR), color='blue', linestyle='dashed', label='Energy')
plt.vlines(R_eq, -1.3, 0, color='r', label='R_eq')
plt.annotate("(R_eq, Min. Energy)", xy=(R_eq, min_energy - 0.1))

# add labels and legend
plt.xlabel('R (a.u)')
plt.ylabel('Energy (Hartrees)')
plt.ylim(-1.3, 0)
plt.grid()
plt.legend(loc='best')

# display plot
plt.show()
```
